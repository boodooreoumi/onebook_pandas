```python
import matplotlib.pyplot as plt
```

```python
# figure : 그림이 그려지는 캔버스
# axes : 하나의 그래프
```

```python
# add_axes를 사용하면 서브플롯의 크기와 위치를 자유롭게 지정가능
```

```python
import seaborn as sns
anscombe=sns.load_dataset('anscombe')
df1=anscombe[anscombe['dataset']=='I']
df2=anscombe[anscombe['dataset']=='II']
df3=anscombe[anscombe['dataset']=='III']
df4=anscombe[anscombe['dataset']=='IV']
```

```python
# figure 객체 생성
fig = plt.figure(figsize=(8,6))
# figure 객체의 add_axes 메소드로 위치와 크기를 지정하여 axes 객체를 생성
ax1=fig.add_axes([0,0.5,0.4,0.4])
ax2=fig.add_axes([0.5,0.5,0.4,0.4])
ax3=fig.add_axes([0,0,0.4,0.4])
ax4=fig.add_axes([0.5,0,0.4,0.4])
# axes에 그래프를 그린다
ax1.plot(df1['x'],df1['y'],'D')
ax2.plot(df2['x'],df2['y'],'D')
ax3.plot(df3['x'],df3['y'],'D')
ax4.plot(df4['x'],df4['y'],'D')
# axes에 제목 추가
ax1.set_title('ax1')
ax2.set_title('ax2')
ax3.set_title('ax3')
ax4.set_title('ax4')
```

```python
# axes를 행,열로 쪼개어 서브플롯 그리기
```

```python
# axes 객체를 행,열로 쪼개어 생성하기
#  축을 공유 통일
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(8,6), sharex=True, sharey=True)
# axes[행번호][열번호] 형태로 접근하여 그래프 그리기
ax[0][0].plot(df1['x'],df1['y'],'D')
ax[0][1].plot(df2['x'],df2['y'],'D')
ax[1][0].plot(df3['x'],df3['y'],'D')
ax[1][1].plot(df4['x'],df4['y'],'D')
# 각 그래프에 제목 추가
ax[0][0].set_title('ax1')
ax[0][1].set_title('ax2')
ax[1][0].set_title('ax3')
ax[1][1].set_title('ax4')
# 각 그래프에 그리드 추가
ax[0][0].grid()
ax[0][1].grid(color='pink')
ax[1][0].grid(color='green')
ax[1][1].grid(color='skyblue')
# 그래프 전체 제목
fig.suptitle('Anscombe', size=20)
# 그래프 간격, 크기 최적화
fig.tight_layout()
```

```python
# 전체 행 열과 그래프 순서에 따라 서브플롯 그리기 = add_subplot
```

```python
# figure객체를 생성
fig=plt.figure(figsize=(9,6))
# 서브플롯을 그릴 axes 객체를 생성
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2, sharex=ax1, sharey=ax1)
ax3=fig.add_subplot(2,2,3, sharex=ax1, sharey=ax1)
ax4=fig.add_subplot(2,2,4)
# axes 객체에 그래프를 그린다.
ax1.plot(df1['x'],df1['y'],'D')
ax2.plot(df2['x'],df2['y'],'D')
ax3.plot(df3['x'],df3['y'],'D')
ax4.plot(df4['x'],df4['y'],'D')
# 틱 변경하기
#ax1.set_xticks(range(1,20,1))
#ax1.set_yticks(range(1,15,1))
# 범례표시하기
ax1.legend(loc=2)
ax2.legend(loc=2)
ax3.legend(loc=2)
ax4.legend(loc=2)
#figure 제목 추가
fig.suptitle('Anscombe', size=20)
# 그래프 크기,간격 최적화
fig.tight_layout()
plt.show()
```
