```python
bins = [0, 6000, 100000, df2['이동거리'].max()]
pd.cut(df2['이동거리'],bins,right=False) # right=false로 지정시 우측범위를 포함하지 않음
```

```python
labels = ['적음','보통','많음']
pd.cut(df2['이동거리'],bins,labels=labels,right=False)
```

```python
pd.cut(df2['이동거리'],bins=3).value_counts() # 고르게 분포한 데이터라면 괜찮지만, 튀는 이상치가 있는 경우 안좋은 결과 초래
```

```python
# 동일한 갯수를 갖도록 구간 분할
pd.qcut(df2['이동거리'], q=3).value_counts()
```

```python
# qcut 또한 범위를 조정가능
qcut_bins=[0,0.2,0.8,1]
pd.qcut(df2['이동거리'], qcut_bins).value_counts()
```

```python
# qcut 역시 labels을 지정가능
qcut_labels = ['적음','보통','많음']
pd.qcut(df2['이동거리'], qcut_bins, labels=qcut_labels).value_counts()
```

```python
sample = sns.load_dataset('titanic')
sample.head(3)
```

- age를 다음과 같은 구간을 가지도록 나누어 주세요
- 0세 초과 ~ 15세 이하
- 15세 초과 ~ 30세 이하
- 30세 초과 ~ 45세 이하
- 45세 초과 ~ 최대값
- 나눈 다음 age_bin 컬럼을 생성하여 나눈 값을 대입하고 분포를 출력하세요
- cut() 함수를 활용하세요

```python
bins = [0, 15, 30, 45, sample['age'].max()]
pd.cut(sample['age'],bins,right=True).value_counts() # right=false로 지정시 우측범위를 포함하지 않음
```

- 나이를 3개의 균등 분할을 갖도록 구간을 나누고 분포를 출력하세요
- 나이가 어린 그룹부터 차례대로 'young', 'normal', 'old' 라벨을 부여하세요
- qcut() 함수를 활용하세요

```python
# 동일한 갯수를 갖도록 구간 분할
pd.qcut(sample['age'], q=3).value_counts()
```

```python
# qcut 역시 labels을 지정가능
qcut_labels = ['young','normal','old']
pd.qcut(sample['age'], q=3, labels=qcut_labels).value_counts()
```
