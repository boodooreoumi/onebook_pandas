- 데이터필터링

```python
# cyl=4인 데이터의 수
cond1 = (df['cyl']==4)
len(df[cond1])
```

```python
len(df[df['cyl']==4])
```

```python
# mpg가 22이상인 데이터수
cond2 = (df['mpg'] >= 22)
len(df[cond2])
```

```python
len(df[df['mpg']>=22])
```

```python
# 2개 조건 필터링
df[cond1 & cond2]
```

```python
print(len(df[cond1 & cond2]))
```

```python
# or
df[cond1 | cond2]
```

```python
print(len(df[cond1 | cond2]))
```

```python
# 한번에 코딩할 경우
print(len(df[(df['cyl']==4) & (df['mpg']>=22)]))
print(len(df[(df['cyl']==4) |(df['mpg']>=22)]))
```

- 데이터변경(조건문)

```python
import numpy as np
df = pd.read_csv('mtcars.csv')
```

```python
# np.where 활용
# hp 변수 값 중에서 205가 넘는 값은 205로 처리하고, 나머지는 그대로 유지
df['hp']=np.where(df['hp']>=205, 205, df['hp'])
```

```python
# 내림차순 정렬
df.sort_values('hp', ascending=False).head(10)
```

- 이상치를  max값이나 min값으로 대체할 경우 조건문 활용
