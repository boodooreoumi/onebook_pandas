```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 그래프에 한글폰트 설정
plt.rcParams['font.family']='Malgun Gothic'
# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus']=False
```

```python
# 그리드 스타일:라인스타일-- 투명도0.3
plt.rcParams['grid.linestyle']='--'
plt.rcParams['grid.alpha']=0.3
```

```python
# 수직선, 수평선 그리기
# 수평선 그리기
plt.plot([1,2,3,4], 'ko')

plt.axhline(2,0,1,color='b',alpha=0.5,ls=':')
plt.hlines(3,0.0,3.0,color='darkorange',ls='-.')

plt.show()
```

```python
# 수직선그리기
plt.plot([1,2,3,4], 'ko')
plt.axvline(1.0,0,1,color='b',alpha=0.5,ls=':')
plt.vlines(2.0,1.0,4.0,color='b',alpha=0.5,ls=':')
plt.show()
```

```python
# 예제 : 요일별 판매 테이블수

import seaborn as sns
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
# 요일별 테이블수
s=tips['day'].value_counts()
s
```

```python
plt.figure(figsize=(15,3))
plt.barh(s.index,s.values)
plt.xticks(range(0,91,1),rotation=90)
plt.grid(axis='x')
plt.axvline(s['Fri'],color='k')
plt.axvline(s['Thur'],color='k')
plt.axvline(s['Sun'],color='k')
plt.axvline(s['Sat'],color='k')
plt.show()
```

```python
# 텍스트 추가하기
```

```python
plt.plot([1,2,3,4], 'ko')
plt.text(2.1,3,'(x:2,y=3)',ha='left',va='bottom', fontsize=12, rotation=45
        ,bbox={'boxstyle':'round','fc':'skyblue','ec':'blue','alpha':0.3})
plt.axhline(3,ls=':',alpha=0.5,lw=0.5)
plt.axvline(2,ls=':',alpha=0.5,lw=0.5)
plt.plot(2,3,'ro')
plt.show()
```

```python
# 화살표와 텍스트 추가하기
plt.plot([1,2,3,4], 'ko')

plt.axhline(2, color='orange', lw=0.5, alpha=0.5)
plt.axvline(1, color='orange', lw=0.5, alpha=0.5)
plt.plot(1,2,'ro')

plt.annotate('(x:1,y:2)',xy=(1,2),xytext=(1.5,2.5)
             ,arrowprops={'width':1,'headwidth':10,'headlength':10,'shrink':0.1}
             ,fontsize=12, color='r')

plt.show()
```
