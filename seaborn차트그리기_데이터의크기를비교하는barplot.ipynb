```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 그래프에 한글폰트 설정
plt.rcParams['font.family']='Malgun Gothic'
# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus']=False
```

```python
# 샘플데이터
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
# 요일별 팁 평균
# matplotlib으로 시각화
# 데이터 가공
```

```python
day_tip_mean=tips.groupby('day')['tip'].mean()
day_tip_mean
```

```python
plt.bar(day_tip_mean.index,day_tip_mean)
plt.xlabel('day')
plt.ylabel('tip')
plt.title('요일별 팁 평균', size=15)
```

```python
# seaborn으로 그리기
sns.barplot(data=tips,x='day',y='tip',ci=None)
plt.title('요일별 팁 평균', size=15)
plt.show()
```

```python
# 요일별 팁 합계
day_tip_sum=tips.groupby('day')['tip'].sum()
day_tip_sum
```

```python
plt.bar(day_tip_sum.index,day_tip_sum)
plt.xlabel('day')
plt.ylabel('tip')
plt.title('요일별 팁 합계', size=15)
plt.show()
```

```python
# seaborn으로 그리기
sns.barplot(data=tips,x='day',y='tip',ci=None, estimator=sum)
plt.title('요일별 팁 합계', size=15)
plt.show()
```

```python
# 요일별 팁합계를 흡연여부로 비교, hue만 추가
```

```python
# seaborn으로 그리기
sns.barplot(data=tips,x='day',y='tip',ci=None, estimator=sum, hue='smoker')
plt.title('요일별 팁 합계', size=15)
plt.show()
```

```python
# hue 색상변경:palette=구분:색상딕셔너리
# seaborn으로 그리기
# 색깔변경하고 그리드 추가
sns.barplot(data=tips,x='day',y='tip',ci=None, estimator=sum, hue='smoker'
           ,palette={'Yes':'gray','No':'skyblue'})
plt.title('요일별 팁 합계', size=15)
plt.grid(axis='y',ls=':')
plt.show()
```
