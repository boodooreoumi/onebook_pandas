```python
df.drop(columns=['환자번호','국적','환자정보','조치사항','이동경로','등록일','수정일','노출여부'],inplace=True)
df
```

```python
# 자료형 확인 및 변환
df.dtypes
```

```python
df['확진일']=pd.to_datetime(df['확진일'])
df.dtypes
```

```python
df['지역'].nunique()
```

```python
df['지역'].unique()
```

```python
df['지역']=df['지역'].str.strip()
```

```python
df['지역'].nunique()
```

```python
df['지역']=df['지역'].astype('category')
```

```python
df.dtypes
```

```python
df.info()
```

```python
# 결측치 분석
df.isnull().sum()
```

```python
# 피봇테이블 만들기
df_gu = pd.pivot_table(df,index='확진일',columns='지역',values='연번',aggfunc='count',margins=True,margins_name='All')
df_gu
```

```python
# 서울시 일별 확진자 동향
s_date=df_gu['All'][:-1]
s_date
```

```python
# 서울시 일별 추가확진자가 많았던 순으로 보기
s_date.sort_values(ascending=False)
```

```python
# 서울시 일별 추가확진자 시각화
x=s_date.index
y=s_date.values
plt.plot(x,y)
plt.title('서울시 일별 추가확진자(2021.09.28)')
plt.xlabel('확진일')
plt.ylabel('추가확진자수')
plt.xticks(rotation=45)
plt.show()
```

```python
# 서울시 구별 추가 확진자 및 접촉력 분석
df_gu
```

```python
s_gu=df_gu.loc['All'][:-1]
s_gu
```

```python
s_gu=s_gu.sort_values(ascending=False)
s_gu
```

```python
# 서울시 구별 누적확진자가 많은 순으로 시각화
x=s_gu.index
y=s_gu.values
plt.figure(figsize=(10,7))
plt.title("서울시 구별 누적확진자(2021.09.28.)",size=20)
plt.barh(x,y)
plt.show()
```

```python
# 최근일 기준 지역별 추가확진자
s_gu=df_gu.iloc[-2][:-1]
s_gu=s_gu.sort_values(ascending=False)
```

```python
x=s_gu.index
y=s_gu.values
plt.figure(figsize=(10,7))
plt.barh(x,y)
plt.show()
```
