```python
# 행을 열로 보내기(피봇)
df=pd.read_csv('data/scores.csv')
df=df.head(2)
df=df.melt(id_vars='name',var_name='subject', value_name='score')

def get_grade(x):
    if x>=90: grade='A'
    elif x>=80: grade='B'
    elif x>=70: grade='C'
    elif x>=60: grade='D'
    else: grade='F'
    return grade

df['grade']=df['score'].apply(get_grade)
df=df.sort_values('name')
df
```

```python
# pivot table : name,subject,score
df.pivot(index='name',columns='subject',values='score')
```

```python
# pivot table : name,subject,grade
df.pivot(index='name',columns='subject',values='grade')
```

```python
# pivot table : name,subject,[score,grade]
df.pivot(index='name',columns='subject',values=['score','grade'])
```

```python
# pivot table : name,subject
df.pivot(index='name',columns='subject')
```

```python
import pandas as pd
df=pd.DataFrame({'item':['shirts','shirts','shirts','shirts','shirts','pants','pants','pants','pants'],
                 'color':['white','white','white','black','black','white','white','black','black'],
                 'size':['small','large','large','small','small','large','small','small','large'],
                 'sale':[1,2,2,3,3,4,5,6,7],
                 'inventory':[2,4,5,5,6,6,8,9,9]})
df
```

```python
# item,size별 재고합계
df.pivot_table(index='item', columns='size', values='inventory', aggfunc='sum')
```

```python
# [item,color],size별 재고합계
df.pivot_table(index=['item','color'], columns='size', values='inventory', aggfunc='sum')
```

```python
# null값은 0으로
df.pivot_table(index=['item','color'], columns='size', values='inventory', aggfunc='sum', fill_value=0)
```

```python
# [item,color],size별 판매,재고합계
df.pivot_table(index=['item','color'], columns='size', values=['sale','inventory'], aggfunc='sum')
```

```python
# 타이타닉호 데이터 불러오기
import pandas as pd
df = pd.read_csv('data/titanic.csv')
df.head()
```

```python
df_titanic =df[['Survived','Pclass','Sex','Age','Embarked']]
df_titanic = df_titanic.dropna()
df_titanic.head()
```

```python
len(df_titanic)
```

```python
# 성별, 객실등급별 생존자수
df_titanic.pivot_table(index="Sex",columns="Pclass",values="Survived",aggfunc="count",margins=True)
```

```python
df_titanic.pivot_table(index="Sex",columns="Pclass",values="Survived",aggfunc="sum",margins=True)
```

```python
# 성별 객실등급별 생존율
df_titanic.pivot_table(index="Sex",columns="Pclass",values="Survived",aggfunc="mean",margins=True)
```

```python
df_titanic.pivot_table(index="Sex",columns="Pclass",values="Survived",margins=True)
```
