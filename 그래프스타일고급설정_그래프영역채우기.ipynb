```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 그래프에 한글폰트 설정
plt.rcParams['font.family']='Malgun Gothic'
# 그래프에 마이너스 기호 깨지는 문제 해결
plt.rcParams['axes.unicode_minus']=False
```

```python
# 그리드 스타일:라인스타일-- 투명도0.3
plt.rcParams['grid.linestyle']='--'
plt.rcParams['grid.alpha']=0.3
# 폰트사이즈 12
plt.rcParams['font.size']=12
```

```python
# 그래프 영역 채우기
# 가로방향 채우기
x=[1,2,3,4]
y=[1,4,3,6]
plt.plot(x,y)
plt.fill_between(x[1:3],y[1:3],color='skyblue',alpha=0.5,hatch='/')
plt.show()
```

```python
# 그래프 영역 채우기
# 세로방향 채우기
x=[1,2,3,4]
y=[1,4,3,6]
plt.plot(x,y)
plt.fill_betweenx(y[:2],x[:2],color='skyblue',alpha=0.5,hatch='/')
plt.show()
```

```python
# 두 그래프 사이의 영역 채우기
x=[1,2,3,4]
y1=[1,3,4,6]
y2=[2,4,7,9]

plt.plot(x,y1)
plt.plot(x,y2)
plt.fill_between(x[1:3],y1[1:3],y2[1:3],color='darkorange',alpha=0.5)

plt.show()
```

```python
# 다각형 채우기
x=[1,2,3,4]
y=[1,4,3,6]
plt.plot(x,y)
plt.grid()

plt.fill([1.75,2.25,2.25,1.75],[3.5,3.5,4.5,4.5], alpha=0.1)
plt.fill([2.75,3.25,3.0],[2.5,2.5,3.5], alpha=0.1)

plt.show()
```
