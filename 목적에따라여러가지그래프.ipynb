```python
# 라이브러리 입포트
import matplotlib.pyplot as plt
```

```python
# "시간"의 시각화 : 어떤아이의 나이별 키의 변화
age=[7,8,9,10,11,12,13,14,15]
height=[122.1,127.9,133.4,138.8,142.7,151.4,155.6,165.0,169.2]
```

```python
# 선그래프
plt.plot(age,height)
plt.show()
```

```python
# '수량비교' 시각화
blood_type=['A','B','O','AB']
count=[48,35,15,5]
```

```python
# 막대그래프
plt.bar(blood_type,count)
plt.show()
```

```python
# 가로막대그래프
plt.barh(blood_type,count)
```

```python
# '비율'의 시각화
blood_type=['A','B','O','AB']
count=[45,35,15,5]
```

```python
plt.pie(count,labels=blood_type)
plt.show()
```

```python
# '분포'의 시각화
# 히스토그램, 상자수염, 바이올린
# 1부터 100사이의 정수 1000개

import numpy as np
data=np.random.randint(1,101,1000)
data

plt.hist(data, bins=100) # 100개로 쪼개어 보여줌
plt.show()
```

```python
# 상자수염 그래프
plt.boxplot(data)
plt.show()
```

```python
# 바이올린 그래프
plt.violinplot(data)
plt.show()
```

```python
# 관계의 시각화
# 지불금액에 따른 팁

import seaborn as sns
tips=sns.load_dataset('tips')[['total_bill','tip']]
tips

plt.scatter(tips['total_bill'],tips['tip'])
plt.show()
```
