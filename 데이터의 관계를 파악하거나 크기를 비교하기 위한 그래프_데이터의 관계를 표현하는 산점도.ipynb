```python
# 데이터
# 크기는 s 색깔을 c
x=[1,5,6,9,10]
y=[1,5,3,9,7]
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y,'D')
```

```python
# 점의 크기 색상
plt.scatter(x,y,s=[100,200,50,60,30], c=['r','b','y','g','k'])
```

```python
# 컬럽맵으로 색상지정
plt.scatter(x,y,c=y, cmap='cool')
plt.colorbar()
```

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
# 지불금액과 팁의 관계
# 지불금액이 클수록 팁도 많이 줄까?
```

```python
plt.scatter(tips['total_bill'],tips['tip'])
```

```python
# 점의 크기로 테이블 인원 표시
plt.scatter(tips['total_bill'],tips['tip'], s=tips['size']*15, alpha=0.5)
```

```python
# 점의 색으로 성별표시
```

```python
# 성별에 따른 colot 컬럼 추가
def set_color(x):
    if x=='Male':
        return 'blue'
    elif x=='Female':
        return 'red'
tips['color']=tips['sex'].apply(set_color)
```

```python
tips.head(5)
```

```python
plt.scatter(tips['total_bill'],tips['tip'], s=tips['size']*15, alpha=0.5, c=tips['color'])
```
