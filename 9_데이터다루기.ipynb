```python
# BeautifulSoupÏùÄ ÌôàÌéòÏù¥ÏßÄ ÏÜåÏä§Î•º Í∞ÄÏ†∏Ïò§Í∏∞Í∞Ä Î∂àÍ∞ÄÌïòÍ∏∞ ÎïåÎ¨∏Ïóê requests Î™®ÎìàÍ≥º Ìï®Íªò ÏÇ¨Ïö©
import requests
from bs4 import BeautifulSoup

req = requests.get("https://naver.com")
html = req.text
soup = BeautifulSoup(html, 'html.parser')

print(soup.title)
print(soup.title.name)
print(soup.title.string)
```

```python
# findÎ•º ÌôúÏö©ÌïòÏó¨ htmlÌÉúÍ∑∏ Ï≤´Î≤àÏß∏ Í∞í Í∞ÄÏ†∏Ïò§Í∏∞Í∞Ä Í∞ÄÎä•
print(soup.find('a'))
```

```python
# find_allÏùÑ ÌôúÏö©ÌïòÎ©¥ HTMLÌÉúÍ∑∏ Î™®Îì†Í∞í Î¶¨Ïä§Ìä∏ ÌòïÌÉúÎ°ú Í∞ÄÏ†∏Ïò§Í∏∞
print(soup.find_all('a')[1])
```

```python
# find,find_allÏòµÏÖòÏúºÎ°ú css ÌòπÏÑ± ÌôúÏö©ÌïòÏó¨ ÏÑ∏Î∂Ä ÌïÑÌÑ∞ÎßÅ
print(soup.find_all('span',class_='blind'))
```

```python
print(soup.find_all('span',attrs={"class": "blind"}))
```

```python
# select_one,select : cssÏÑ†ÌÉùÏûêÎ•º ÌôúÏö©ÌïòÏó¨ ÏõêÌïòÎäî Îç∞Ïù¥ÌÑ∞ Í∞ÄÏ†∏Ïò§Í∏∞
print(soup.select_one('a'))
```

```python
print(soup.select('a'))
```

```python
# get_textÎäî ÌôàÌéòÏù¥ÏßÄ ÏÜåÏä§ ÎÇ¥ ÌÉúÍ∑∏ Ï†úÏô∏ ÌÖçÏä§Ìä∏Î•º Ï∂úÎ†•
text = soup.find("span", attrs={"class":"blind"})
print(text)
print(text.get_text())
```

```python
# reÎ™®Îìà
# Ï†ïÍ∑úÌëúÌòÑÏãùÏù¥ÎûÄ ÌäπÏ†ïÌïú Í∑úÏπô Î¨∏ÏûêÏó¥ ÌëúÌòÑÏù¥ÎÇò Î¨∏ÏûêÏó¥ Í≤ÄÏÉâ Î∞è ÏπòÌôòÏóê ÌôúÏö©
# Îç∞Ïù¥ÌÑ∞Î∂ÑÏÑù Î∞è Í∞ÄÍ≥µÏù¥ Ïö©Ïù¥ÌïòÏó¨ ÏõêÌïòÎäî Îç∞Ïù¥ÌÑ∞ Ï∂úÎ†•

import re

text = ('111<head>ÏïàÎÖïÌïòÏÑ∏Ïöî</head>22')
body = re.search('<head.*/head>',text)

body = body.group()
print(body)
```

```python
# Î©îÌÉÄÎ¨∏ÏûêÎ•º ÌôúÏö©ÌïòÏó¨ Í∑úÏπô Ï†ïÏùò
import re
data = """<head>ÏïàÎÖïÌïòÏÑ∏Ïöî!!<title> ÌÜµÍ≥ÑÎ•º ÏúÑÌïú ÌååÏù¥Ïç¨ Îç∞Ïù¥ÌÑ∞Î∂ÑÏÑù Í∏∞Ï¥à Í≥ºÏ†ïÏûÖÎãàÎã§. </title> ÏàòÍ≥†ÌïòÏÖ®ÏäµÎãàÎã§. </head>"""
data = re.search('<title.*/title>',data)
data = data.group()
print(data)

data = re.sub('<.+?>','',data)
print(data)
```

```python
# 2.  split() ÌôúÏö©ÌïòÏó¨ ÏõêÌïòÎäê Î∂ÄÎ∂ÑÏùÑ Î∂ÑÎ¶¨ÌïòÏó¨ Î¶¨Ïä§Ìä∏Ïóê Ï†ÄÏû•
data = req.text
data = data.split('<!-- Îû≠ÌÇπÎ¶¨Ïä§Ìä∏ -->')[1]
data = data.split('<!-- ÌèâÏ†êÏàúÏùº Îïå ÌèâÏ†ê Ï∂îÍ∞ÄÌïòÍ∏∞ -->')
```

```python
# bs4 ÌôúÏö©
import requests
import bs4
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
req = requests.get(url)
data = req.text
soup = bs4.BeautifulSoup(data, 'html.parser')
data = soup.find_all('div', class_='tit3')
```

```python
for i in data:
    print(i)
```

```python
for i in data:
    print(i.get_text())
```

```python
for i in data:
    print(i.get_text().strip())
```

```python
# ÌÜµÍ≥ÑÍµêÏú°Ïõê Í≥µÏßÄÏÇ¨Ìï≠ Í∞ÄÏ†∏Ïò§Í∏∞
import requests
import bs4

url = "https://sti.kostat.go.kr/coresti/site/board/BoardList.do?gmenu=3&rmenu=04&cmenu=0401&flag=nots"
req = requests.get(url)
data = req.text
print(data)
```

```python
# bs4Î™®ÎìàÌôúÏö© - htmlÌÉúÍ∑∏Í∏∞Î∞ò ÌååÏã± - classÏÜçÏÑ±Ïù¥ table-a1Ïù∏ Í≤ÉÏùÑ Í∞ÄÏ†∏Ïò¥
data = bs4.BeautifulSoup(data,'html.parser')
data=data.find('table',class_='table-a1')
data=data.find('tbody')
data=data.find('tr')
data=data.find('td',class_='text_left')
print(data.get_text())
```
