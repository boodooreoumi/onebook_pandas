```python
#bs4
# 홈페이지 내 데이터 추출에 도움
# HTML 태그 파서 활용하여 사용하기 편한 파이썬 객체를 제공
# 웹 문서의구조를 이해하면 원하는 데이터 추출이 용이
# 모듈 내 BeautifulSoup() 클래스를 이용하여 HTML 문서,parser를 전달하여 분석 결과 객체를 저장
# parser에 따라 태그 추가,무시,강제변경의 작업을 수행
```

```python
# BeautifulSoup은 홈페이지 소스를 가져오기가 불가하기 때문에 requests 모듈과 함께 사용
import requests
from bs4 import BeautifulSoup

req = requests.get("https://naver.com")
html = req.text
soup = BeautifulSoup(html, 'html.parser')

print(soup.title)
print(soup.title.name)
print(soup.title.string)
```

```python
# find를 활용하여 html태그 첫번째 값 가져오기가 가능
print(soup.find('a'))
```

```python
# find_all을 활용하면 HTML태그 모든값 리스트 형태로 가져오기
print(soup.find_all('a')[1])
```

```python
# find,find_all옵션으로 css 혹성 활용하여 세부 필터링
print(soup.find_all('span',class_='blind'))
```

```python
print(soup.find_all('span',attrs={"class": "blind"}))
```

```python
# select_one,select : css선택자를 활용하여 원하는 데이터 가져오기
print(soup.select_one('a'))
```

```python
print(soup.select('a'))
```

```python
# get_text는 홈페이지 소스 내 태그 제외 텍스트를 출력
text = soup.find("span", attrs={"class":"blind"})
print(text)
print(text.get_text())
```

```python
# re모듈
# 정규표현식이란 특정한 규칙 문자열 표현이나 문자열 검색 및 치환에 활용
# 데이터분석 및 가공이 용이하여 원하는 데이터 출력

import re

text = ('111<head>안녕하세요</head>22')
body = re.search('<head.*/head>',text)

body = body.group()
print(body)
```

```python
# 메타문자를 활용하여 규칙 정의
import re
data = """<head>안녕하세요!!<title> 통계를 위한 파이썬 데이터분석 기초 과정입니다. </title> 수고하셨습니다. </head>"""
data = re.search('<title.*/title>',data)
data = data.group()
print(data)

data = re.sub('<.+?>','',data)
print(data)
```

```python
# 데이터 수집 및 가공
# 1. N사 영화페이지 영화랭킹 가져오기
# 1. 'https://movie.naver.com/movie/sub/rank/rmovie.nhn'
```

```python
import requests
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
req = requests.get(url)
print(req.text)
```

```python
# 2.  split() 활용하여 원하느 부분을 분리하여 리스트에 저장
data = req.text
data = data.split('<!-- 랭킹리스트 -->')[1]
data = data.split('<!-- 평점순일 때 평점 추가하기 -->')
```

```python
# 3. 반복문을 활용해서 원하는 데이터를 추출
for i in data:
    print(i.split('title=')[1].split('"')[0])
```

```python
# bs4 활용
import requests
import bs4
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
req = requests.get(url)
data = req.text
soup = bs4.BeautifulSoup(data, 'html.parser')
data = soup.find_all('div', class_='tit3')
```

```python
for i in data:
    print(i)
```

```python
for i in data:
    print(i.get_text())
```

```python
for i in data:
    print(i.get_text().strip())
```
