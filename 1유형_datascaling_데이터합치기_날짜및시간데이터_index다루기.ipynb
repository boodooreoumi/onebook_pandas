- 날짜,시간데이터, index 다루기

```python
df=pd.DataFrame({
    '날짜':['20230105','20230105','20230223','20230223','20230312','20230422','20230511'],
    '물품':['A','B','A','B','A','B','A'],
    '판매수':[5,10,15,15,20,25,40],
    '개당수익':[500,600,500,600,600,700,600]})
df
```

```python
df.info()
```

```python
# 데이터 타입 datetime으로 변경
df['날짜']=pd.to_datetime(df['날짜'])
df.info()
```

```python
# 년, 월, 일 변수(열) 추가하기
df['year']=df['날짜'].dt.year
df['month']=df['날짜'].dt.month
df['day']=df['날짜'].dt.day
df
```

```python
# 날짜 구간 필터링
df[df['날짜'].between('2023-01-01','2023-02-23')]  #좌우 모두 포함

# between('2023-01-05 12:00:00', '2023-01-05 12:40:00')
```

```python
df=pd.DataFrame({
    '날짜':['20230105','20230105','20230223','20230223','20230312','20230422','20230511'],
    '물품':['A','B','A','B','A','B','A'],
    '판매수':[5,10,15,15,20,25,40],
    '개당수익':[500,600,500,600,600,700,600]})

# 데이터타입 날짜로 변경
df['날짜'] = pd.to_datetime(df['날짜'])

df=df.set_index('날짜') # drop=True(디폴트) or False
df.head()
```

```python
print(df.loc['2023-01-05':'2023-02-23'])
print(df.loc[(df.index>='2023-01-05')&(df.index<='2023-02-23')])
```

- 시간 다루기

```python
df=pd.DataFrame({
    '물품':['A','B','A','B','A','B','A'],
    '판매수':[5,10,15,15,20,25,40],
    '개당수익':[500,600,500,600,600,700,600]})

time = pd.date_range('2023-09-24 12:25:00','2023-09-25 14:45:30', periods=7)
df['time']=time

df=df[['time','물품','판매수','개당수익']]
```

```python
# 시간이 변수로 있을 경우(between 함수)

df = df[df['time'].between('2023-09-24 12:25:00','2023-09-24 21:11:50')]
print(df)
```

```python
df=pd.DataFrame({
    '물품':['A','B','A','B','A','B','A'],
    '판매수':[5,10,15,15,20,25,40],
    '개당수익':[500,600,500,600,600,700,600]})

time = pd.date_range('2023-09-24 12:25:00','2023-09-25 14:45:30', periods=7)
df['time']=time

df=df[['time','물품','판매수','개당수익']]

# 시간이 index로 있을 경우(between_time, loc함수)
# 주의 : 시간이 index에 위치해야 함

df=df.set_index('time')

df=df.between_time(start_time='12:25:00', end_time='21:11:50')
print(df)
```

```python
# loc 함수 활용
# 주의 : 시간이 index에 위치해야 함

print(df.loc['2023-09-24 12:25:00':'2023-09-24 21:11:50'])
print(df.loc[(df.index>='2023-09-24 12:25:00')&(df.index<='2023-09-24 21:11:50')])
```
