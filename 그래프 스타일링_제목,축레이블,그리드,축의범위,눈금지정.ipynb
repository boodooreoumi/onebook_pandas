```python
# 샘플데이터
# 월별 몸무게 데이터프레임 만들기
import pandas as pd
df=pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[80,70,75,73,70]})
df
```

```python
# 그래프 그리기
x=df['월']
y=df['몸무게']
plt.plot(x,y,'ro-',mec='b',mfc='b',lw=3,ms=7)
plt.title("월별 몸무게 변화", loc='right',pad=10, color='r',fontsize=20
          , fontweight='bold',fontfamily='Malgun Gothic')
plt.xlabel('월',loc='right', color='b',fontsize=12) # left,right
plt.ylabel('몸무게(kg)',loc='top',labelpad=10,fontsize=12) #bottom,center,top
plt.grid(axis='x',color='skyblue',alpha=0.5,ls=':') # axis='y' y축만 그리드
plt.show()
```

```python
# 여러개의 그래프와 범례
```

```python
df1 = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[80,78,75,73,70]})
df2 = pd.DataFrame({'월':[1,2,3,4,5],'몸무게':[60,62,59,55,54]})
```

```python
df1
```

```python
df2
```

```python
# 여러 개의 그래프 각각 그리기
plt.plot(df1['월'],df1['몸무게'])
plt.show()

plt.plot(df2['월'],df2['몸무게'])
plt.show()
```

```python
# x축 틱 라벨
def fn_xtick(x):
    x = str(x)
    if x=='0' or x=='13':
        return ''
    else:
        return x+'월'
    
xtick_label=pd.Series(range(0,14,1))
xtick_label=xtick_label.apply(fn_xtick)
xtick_label
```

```python
def fn_ytick(x):
    x=str(x)
    return x+'kg'

ytick_label=pd.Series(range(30,95,5))
ytick_label=ytick_label.apply(fn_ytick)
ytick_label
```

```python
# 두 그래프 한번에 그리기
plt.plot(df1['월'],df1['몸무게'])
plt.plot(df2['월'],df2['몸무게'])

# 그래프제목
plt.title("월별 몸무게 변화", size=15)

# 축라벨
plt.xlabel('월',fontsize=12)
plt.ylabel('몸무게',fontsize=12)

# 틱
plt.xticks(range(0,14,1), labels=xtick_label)
plt.yticks(range(30,95,5), labels=ytick_label)

# 그리드
plt.grid(ls=':')

plt.show()
```

```python
# 두 그래프 한번에 그리기 + 범례
plt.plot(df1['월'],df1['몸무게'], label='James')
plt.plot(df2['월'],df2['몸무게'], label='Amy')
plt.legend(loc='upper right', ncol=2)

# 그래프제목
plt.title("월별 몸무게 변화", size=15)

# 축라벨
plt.xlabel('월',fontsize=12)
plt.ylabel('몸무게',fontsize=12)

# 틱
plt.xticks(range(0,14,1), labels=xtick_label)
plt.yticks(range(30,95,5), labels=ytick_label)

# 그리드
plt.grid(ls=':')

plt.show()
```
