```python
# 데이터 불러오기
import pandas as pd
df = pd.read_csv("data/scores.csv")
df.head()
```

```python
# 열추가하기
df['no']= range(1,len(df)+1)
df.tail()
```

```python
df['sum']=df['kor']+df['eng']+df['math']
df.tail()
```

```python
# 열 수정하기
df['no']=df['no']+100
df.head()
```

```python
# 열 삭제하기
df.drop(columns=['no','sum'], inplace=True)
df.head()
```

```python
# 전체 컬럼명 바꾸기
df.columns = ['이름','국어','영어','수학']
df.head()
```

```python
# 특정 컬럼명 변경하기
df.rename(columns={'이름':'성명'}, inplace=True)
df.head()
```
