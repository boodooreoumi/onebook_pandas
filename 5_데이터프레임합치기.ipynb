```python
# 여러파일로 나눠져 있는 데이터프레임을 연결/병합/결합하는 법에 대해 알아보자

# 구성형태와 속성이 같은 데이터프레임 연결 -> concat

# 여러 데이터프레임 결헙 -> merge : 키가 되는 열 또는 행을 기준으로 데이터를 병합하는 명령
# key가 되는 변수명이 서로 다른 경우의 병합 -> left_on, right_on 을 이용

# 하나의 데이터ㅍ레임에 또 다른 데이터프레임을 결합시키는 명령 -> join
```

```python
import pandas as pd
df1 = pd.read_excel('price.xlsx', index_col='id')
print(df1)
```

```python
df2 = pd.read_excel('valuation.xlsx', index_col='id')
print(df2)
```

```python
df3 = df1.join(df2)
print(df3)
```

```python
df4 = df1.join(df2, how='right')
print(df4)
```

```python
df5 = df1.join(df2, how='inner')
print(df5)
```

```python
df6 = df1.join(df2, how='outer')
print(df6)
```

```python
# 데이터분석 실습
# step1 - 지역별(수도권, 비수도권 자료를 불러와 병합)
import pandas as pd

df2 = pd.read_csv('월간_매매가격지수_아파트_21년_수도권.csv')
df3 = pd.read_csv('월간_매매가격지수_아파트_21년_수도권외.csv')
```

```python
print(df2)
```

```python
print(df3)
```

```python
# 지역별(수도권,비수도권) 자료를 불러와 병합
df5 = pd.concat([df2,df3], ignore_index=True)
print(df5)
```

```python
# 20년 전국아파트 매매개가격지수 자료를 불러와 df1에 할당
# 필요한 자료는 9월부터 12월까지의 자료가 필요
df1 = pd.read_csv('월간_매매가격지수_아파트_20년_전체.csv')
print(df1)
```

```python
# 20년 9월 ~ 12월에 해당하는 자료만 추출하여 df3에 저장
df1_1 = df1.iloc[:,:4]
df1_2 = df1.iloc[:,12:]
df1_3 = pd.concat([df1_1, df1_2], axis=1)
df1_3.info()
```

```python
merge_1 = pd.merge(df1_3.iloc[:,4:], df5, how='outer', on='id')
print(merge_1)
```

```python
merge_1.info()
```

```python
print(merge_1)
```

```python
# 수도권과 수도권외 지역에 대한 최근 1년간 매매지수 평균값 산출
mask1 = merge_1['지 역'].isin(['서울','인천','경기'])
data_1 = merge_1.loc[mask1,:]

mask2 = ~merge_1['지 역'].isin(['서울','인천','경기'])
data_2 = merge_1.loc[mask2,:]
```

```python
M1 = data_1.mean()
M2 = data_2.mean()

DF = pd.DataFrame([M1,M2], index=['수도권','수도권외'])
print(DF)
```
