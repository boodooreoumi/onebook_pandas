```python
# 데이터준비하기
import pandas as pd
df = pd.read_csv("data/scores.csv")
df_copy=df.copy()
```

```python
df_copy.head()
```

```python
df_copy['math']= df_copy['math']+5
df_copy.head()
```

```python
# 점수가 100은 넘을 수 없다. 함수로~~~~
def plus5(x):
    score = x + 5
    if score > 100:
        score = 100
    return score

df_copy['math'] =df['math'].apply(plus5)
df_copy
```

```python
def plusn(x,n):
    score = x + n
    if score > 100:
        score=100
    return score

df_copy['eng']=df_copy['eng'].apply(plusn, n=1)
df_copy
```

- 함수로 행/열의 데이터 집계하기
- axis=0 열단뒤
- axis=1 행단위

```python
df=pd.read_csv("data/scores.csv")
df.index = df['name']
df.drop(columns=['name'], inplace=True)
df_copy=df.copy()
df_copy.head()
```

```python
# 함수만들기
def print_me(x):
    print(x)

# 행 단위 데이터 출력
df_copy.apply(print_me, axis=1)
```

```python
# 합계구하기 함수
def get_sum(x):
    return x.sum()

df_copy['sum']=df_copy.apply(get_sum, axis=1)
df_copy
```

```python
# 과목별 합계
df_copy.apply(get_sum, axis=0)
```

```python
df_copy.loc['sum']=df_copy.apply(get_sum,axis=0)
df_copy.tail()
```
