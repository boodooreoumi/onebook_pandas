```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터준비
x = ['a','b','c','d','e']
y = [1,3,2,10,7]
```

```python
plt.plot(x,y)
plt.show()
```

```python
# 막대그래프
plt.bar(x,y)
plt.show()
```

```python
# 가로막대그래프
plt.barh(x,y)
plt.show()
```

```python
# 산점도
plt.scatter(x,y)
plt.show()
```

```python
# 세가지 그래프 한꺼번에 표시
plt.scatter(x,y,label='scatter')
plt.bar(x,y,label='bar')
plt.plot(x,y,label='plot')
plt.title('Test Graph', size=15)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
