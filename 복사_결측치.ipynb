# 본문

```python
# 모듈
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
# copy()를 하면 복사한 데이터프레임을 수정해도 원본에는 영향을 미치지 않음
df_copy = df.copy()
df_copy.head()
```

```python
# 복사한 데이터의 임의수정
df_copy.loc[0,'age'] = 9999
df_copy.head()
```

```python
df.head()
```

```python
# 결측치
# 1. 결측 데이터 확인
# 2. 결측치가 아닌 데이터 확인
# 3. 결측 데이터 채우기
# 4. 결측 데이터 제거하기
```

```python
# 결측치 확인
# isnull().sum()
df_copy.isnull().sum()
```

```python
# DataFrame 전체 결측 데이터의 갯수
df_copy.isnull().sum().sum()
```

```python
# 결측치가 아닌 데이터 화인
# notnull()
```

```python
df_copy.notnull().sum()
```

```python
# 결측데이터필터링
df_copy.loc[df_copy['age'].isnull()]
```

```python
# 결측치 채우기
df_copy['age'].fillna(700).tail()
```

```python
# 위의 코드는 결과값일뿐 실제값은 변경되지 않았으므로,
# df_copy['age']=df_copy['age'].fillna(700) 으로 해야 한다.
```

```python
# 통계값으로 채우기
# df_copy.loc[0,'age']=22
# 평균값
df_copy['age'].fillna(df_copy['age'].mean()).tail()
```

```python
# 중앙값
df_copy['age'].fillna(df_copy['age'].median()).tail()
```

```python
# 최빈값
# df_copy['deck'].head()
# df_copy['deck'].mode()[0]
df_copy['deck'].fillna(df_copy['deck'].mode()[0]).tail()
```

```python
# NaN 값이 있는 데이터 제거하기 (dropna)
# 한개라도 NaN이 있는 행은 제거가 됨
df_copy.dropna(how='any')
```

```python
# any: 1개 라도 NaN값이 존재시 drop
# all: 모두 NaN값이 존재시 drop
```

```python
df_copy.dropna(how='all')
```

# 연습문제

-1 1. 타이타닉 승객 데이터에서 age가 결측치인 승객의 나이를 30세로 일괄 채워 주세요

```python
df_copy=df.copy()
```

```python
df_copy.loc[df_copy['age'].isnull(), 'age']=30
```

```python
df_copy.isnull().sum()
```

- 남자 승객의 나이의 결측치는, 남자 승객의 나이의 평균으로 채웁니다.
- 여자 승객의 나이의 결측치는, 여자 승객의 나이의 평균으로 채웁니다.
- 결측치를 모두 채운 후 age 컬럼의 평균을 출력합니다.(검증코드 실행하여 올바른 값이 출력 되는지 확인)

```python
df_copy=df.copy()
```

```python
df_copy.loc[df['sex']=='male']['age'].isnull().sum()
```

```python
# 성별이 남성인 나이컬럼만
df_copy.loc[df_copy['sex']=='male','age']
```

```python
# 성별이 남성인 나이컬럼의 평균
df_copy.loc[df_copy['sex']=='male','age'].mean()
```

```python
# 남자승객의 나이에 결측치가 있는 값을
# 남자승객의 나이의 평균으로 대체
df_copy.loc[df_copy['sex']=='male','age'].fillna(df_copy.loc[df_copy['sex']=='male','age'].mean())
```

```python
# 성별이 여성인 나이컬럼만
df_copy.loc[df_copy['sex']=='female','age']
```

```python
df_copy.loc[df_copy['sex']=='female','age'].fillna(df_copy.loc[df_copy['sex']=='female','age'].mean())
```
